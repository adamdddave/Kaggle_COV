```python
import pandas as pd
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
import wordcloud
import numpy as np
```

```python
df = pd.read_csv("../2020-03-13/all_sources_metadata_2020-03-13.csv")
```

```python
df = df[df['abstract'].notnull()]
```

```python
df['abstract'] = df.apply(lambda x : x.abstract.replace(".","").replace(",","").lower().split(),axis=1)
```

```python
from collections import Counter
z = Counter()
#print(z)
for _,row in df.iterrows():
    #print(row)
    z+=Counter(row['abstract'])
#print(z)
```

```python
#make a function to do the frequency map
#https://amueller.github.io/word_cloud/auto_examples/frequency.html#sphx-glr-auto-examples-frequency-py
def makeImage(text):
    wc = wordcloud.WordCloud(background_color="black", max_words=100000000)
    # generate word cloud
    wc.generate_from_frequencies(text)
    # show
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.show()
    plt.savefig("test.pdf")
```

```python

makeImage(z)
```
